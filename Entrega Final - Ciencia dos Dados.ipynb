Neste projeto, será abordado o estudo dos paises da NAFTA (_North_ _American_ _Free_ _Trade_ _Agreement_ ). O bloco econômico é um acordo econômico e comercial formado por Estados Unidos, Canadá e México. Criado efetivamente em 1° de janeiro de 1994, tem como objetivo o fortalecimento das relações comerciais entre esses países. A estratégia estadunidense de criação do bloco visa enfrentar a concorrência dos mercados europeu e especialmente o asiático, que tem tido vigorosa evolução nos últimos anos.




Foram escolhidos para a analise os paises do NAFTA, paises da União Européia, Alemanha, França e Italia, pois os 3 paises são grandes potências européias, e da asia, os 3 paises mais impactantes dos Tigres Asiaticos, China, Japão e Coréia do Sul.

### **Pergunta:**

O plano estadounidense de criar um bloco economico para enfrentar a concorrência dos mercados europeus e asiaticos realmente funcionou?


### **Hipotese:**

A Criacao da NAFTA influenciou a taxa de importacao e exportacao e consequentemente o GDP estadounidense.


### **Mecanismo:**

Analise dos dados de importação, exportação e GDP, antes da criação do NAFTA e depois da criacao do mesmo para os paises, tanto europeus, quanto asiaticos.

### **Analise:**

A NAFTA foi um plano que influenciou muito na importação e exportação estadounidense, com uma média de 4.125^11 de importação  e 4.225^11 de exportação entre os anos de 1980 à 1994, o bloco econômico fez com que em 19 anos, essas taxas subissem para 1.5^12 na importação e 0.975^12 na exportação. Um fato curioso, é que mesmo com esse gigante aumento nas taxas citadas anteriormente, o GDP não foi influenciado.

 Outro ponto, é a disputa de mercado com a China e a Europa. O plano de criação do NAFTA ajudou a economia americana, porém não solucionou o problema, apenas amenizou-o. Referente aos paises europeus, os EUA tiveram um maior sucesso com o plano, pois a diferença de importação e exportação entre os paises aumentou relativamente. Mas em relação a disputas econômicas com a China, não pode-se dizer o mesmo. Com os dados de antes da criação da NAFTA, a venda e compra de produtos fora do pais, já era maior que qualquer pais do Tigre Asiatico, porém com o passar do tempo e a ascensão mercantil chinesa, fez com que as taxas de venda chinesa, depois de 2011, ficassem maiores que a americana.
